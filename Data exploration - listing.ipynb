```python
#shape = data.shape
#print(shape)

print "Our data set has " + str(len(df.index)) + " observations."
print "Our data set has " + str(len(df.columns)) + " predictive variables."

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

#Scatter plot of our data

df_entire = df.loc[df['room_type'] == 'Entire home/apt']
df_priv = df.loc[df['room_type'] == 'Private room']
df_shar = df.loc[df['room_type'] == 'Shared room']

longitude_entire = df_entire['longitude']
longitude_priv = df_priv['longitude']
longitude_shar = df_shar['longitude']

latitude_entire = df_entire['latitude']
latitude_priv = df_priv['latitude']
latitude_shar = df_shar['latitude']


ax.scatter(longitude_entire, latitude_entire, color='b', label='Entire home/apartment')
ax.scatter(longitude_priv, latitude_priv, color='r', label='Private room')
ax.scatter(longitude_shar, latitude_shar, color='g', label='Shared room')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('AirBnb listing')
ax.legend()


plt.show()

df.head(1)

```
